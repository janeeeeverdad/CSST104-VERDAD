**SIMPLE LINEAR REGRESSION: STUDENT'S GRADES**

1. Import Libraries

```python
import pandas as pd #Data Manipulation
import numpy as np #Mathematical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Data Modelling/Training Data
```

2. Loading Dataset

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
```

3. DataFrame

```python
X = ('Hours_of_Study')
Y = ('Grades')
```

4. Visualizing Data

```python
plt.scatter(data['Hours_of_Study'], data['Grades'])
plt.xlabel('Hours of Study')
plt.ylabel('Grades')
plt.title('Grades depending on the hours of study')
plt.grid(True)
plt.show()
```

5. Preparing Data for Modeling

```python
X = data[['Hours_of_Study']]
Y = data[['Grades']]
```

6. Fitting the Linear Regression Model

```python
model = LinearRegression()
model.fit(X,Y)
```

7.  Getting Model Coefficient

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Slope: ", slope)
print("Intercept: ", intercept)
```

8. Visualizing Linear Regression

```python
plt.scatter(X, Y)
plt.plot(X, model.predict(X), color= 'black')
plt.xlabel('Hours of Study')
plt.ylabel('Grades')
plt.title('Linear Regression')
plt.grid(True)
plt.show()
```

9. Making Predictions

```python
new_X = [[10]]
predicted_Y = model.predict(new_X)
print('New Hours of Study', new_X)
print('Predicted Grades for new Hours of study: ', predicted_Y)
```
