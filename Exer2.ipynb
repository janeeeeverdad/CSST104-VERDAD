1. Creating Dataset

```python
import pandas as pd

data = {
    'student id' : range(1, 16),
    'gender' : ['Female', '', 'Female', 'Male','Female', 'Male', 'Female', 'Male', 'Female','Male', 'Female','Male','Female','Male','Female'],
    'test_date': pd.date_range(start = '2024-02-14', periods = 15),
    'study_hours' : [5,2,5,3,6,5,1,4,3,4,2,6,4,5,3],
    'test_score' : [80,75,87,98,75,83,87,90,93,90,83,87,90,94,90],
    'attendance' : ['Present','Absent', 'Absent', 'Absent', 'Present','Present', 'Present', 'Absent', 'Present', 'Absent','Present', 'Absent','Present', 'Absent','Present'],
}

df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index=False)
```

2. Data Loading and Initial Examination

```python
df = pd.read_csv('student_performance.csv')

print (df.head(15))
print (df.isnull().sum())
```

3. Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score', 'study_hours']].describe()

plt.figure(figsize = (12, 5))

#Histogram
plt.subplot(1,2,1)
sne.histplot(df['test_score'], bins = 15, kde=True)
plt.title('Distribution of Test Scores')

#Histogram
plt.subplot(1,2,2)
sne.histplot(df['study_hours'], bins = 15, kde=True)
plt.title('Distribution of Study Hours')

#Scaterplot
plt.figure(figsize = (5,4))
sne.scatterplot(data=df, x='study_hours', y='test_score')
plt.title('Scatter Plot of the Study Hours vs. Test Scores')

```

4. Data Visualization

```python
#Gender Distribution
plt.figure(figsize = (6, 4))
sne.countplot(data=df, x='gender')
plt.title('Distribution of Gender')

#Test Scores Over Time
plt.figure(figsize = (16, 6))
sne.lineplot(data=df, x='test_date', y='test_score')
plt.title('Trend in Test Scores Over Time')

#Test Scores Over Time
plt.figure(figsize = (16, 6))
sne.lineplot(data=df, x='test_date', y='test_score')
plt.title('Trend in Test Scores Over Time')

correlation_matrix = df[['study_hours', 'test_score', 'attendance']].corr()
plt.figure(figsize = (8,6))
sne.heatmap(correlation_matrix, annot=True, cmap='cool')
plt.title('Correlation Heatmap')
```
